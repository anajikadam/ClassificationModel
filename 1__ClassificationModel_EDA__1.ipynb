### Inter Quartile Range
IQR = 75th percentile - 25th percentile <br>
<br>
Lower Bound = q1–1.5*IQR <br>
Upper Bound = q3+1.5*IQR <br>

```python
#FUNCTION TO IDENTIFY OUTLIERS USING IQR METHOD
def iqr(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    min_ = q1 - 1.5 * iqr
    max_ = q3 + 1.5 * iqr
    return col[((col < min_) | (col > max_))]
```

###  Z-score method

Z-Score = (X-mean) / Standard deviation

For numeric columns using pandas to check normality in Python:

- Descriptive statistics.
- Histogram.
- Q-Q plot.
- Shapiro-Wilk test.

```python

import numpy as np
import statsmodels.api as sm
import pylab as py


sm.qqplot(df["precinct"], line ='45')
py.show()
```

```python
from scipy.stats import shapiro

stat, p = shapiro(df["precinct"].sample(5000))

print("Statistics:", stat)
print("p-value:", p)
```

The Shapiro-Wilk test is a statistical test that can be used to assess the normality of a distribution. The null hypothesis of the Shapiro-Wilk test is that the data is normally distributed. If the p-value is less than the significance level, then we can reject the null hypothesis and conclude that the data is not normally distributed.

The following are some of the common distributions of a numeric column:

Normal distribution:

This is the most common distribution. The data is evenly distributed around the mean.
- Exponential distribution: This distribution is used to model events that occur randomly in time. The data is skewed to the right.
- Gamma distribution: This distribution is used to model data that has a positive skew.
- Uniform distribution: This distribution is used to model data that has a constant probability of occurring.

Once you have assessed the normality of your data, you can use this information to make decisions about how to analyze your data.

For example, if your data is normally distributed, you can use parametric tests such as the t-test and the ANOVA.

If your data is not normally distributed, you can use non-parametric tests such as the Mann-Whitney U test and the Kruskal-Wallis test.

```python
#FUNCTION TO DETECT OUTLIERS USING Z-SCORE METHOD
def zscore_outlier(col,lb,ub):
    zscore = ((col - col.mean()) / col.std()).copy()
    return col[((zscore < lb) | (zscore > ub))]
```

```python
zscore_outlier(df.precinct,-3,3)
```
